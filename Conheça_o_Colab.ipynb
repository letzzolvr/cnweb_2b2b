# Formula da Velocidade Média

```
def calcVelMedia(deltaS, deltaT):
  velMedia = (deltaS / deltaT)
  return velMedia
```

teste de serviço: velocidade media

```
print('Velocidade Média: ',calcVelMedia(50, 4))
```

# Fórmula da Aceleração

```
def calcAcel(deltaV, deltaT):
  Acel = (deltaV / deltaT)
  return Acel
```

teste de serviço: aceleração

```
print('Aceleração: ',calcAcel(45, 4))
```

# Fórmulas do Movimento Uniforme (MU)

## Função horária do deslocamento

```
import math
def calcMovUniDesloc(Szero, velocidade, deltaT):
 s = Szero + velocidade
 vFinal = s * deltaT
 res = round(math.sqrt(vFinal), 2)
 return  res
```

teste de serviço: função horária de deslocamento

```
print('Função horária de deslocamento', calcMovUniDesloc(10, 5, 4))
```

# Fórmulas do Movimento Uniformemente Variado (MUV)

## Função horária da posição

```
import math
def calcMovUniPosi(Szero, Vzero, temp, acel):
  s = Szero + Vzero * temp + (1/2) * acel * (temp ** 2)
  return s
```

teste de serviço: Função horária da posição

```
print('Função horária da posição', calcMovUniPosi(10, 5, 2, 3))
```

## Equação de Torricelli

```
import math
def calcTorricelli(vZero, acel, desloc):
  t = 2 * acel * desloc
  vFinal = t + vZero
  res = round(math.sqrt(vFinal))
  return res
```

teste de serviço: equação torricelli

```
print('Equação Torricelli', calcTorricelli(5, 7, 60))
```

# Fórmula do movimento vertical

## Equação de Torricelli para lançamento vertical

```
def calcTorricelliVer(Vzero, grav, alt):
  vFinal1 = (Vzero ** 2) - (2 * grav * alt)
  vFinal2 = (Vzero ** 2) + (2 * grav * alt)
  return vFinal1, vFinal2
```

teste de serviço: equação de torricelli para lançamento vertical

```
print("Torricelli Vertical [Subindo], [Descendo]", calcTorricelliVer(50, 10, 100))
```

## Velocidade vertical em função do tempo

```
def calcVelFunTemp(Vzero, grav, temp):
  vFinal1 = Vzero + (grav * temp)
  vFinal2 = Vzero - (grav * temp)
  return vFinal1, vFinal2
```

teste de serviço: velocidade vertical em função do tempo

```
print('Velocidade vertical em função do tempo', calcVelFunTemp(50, 13, 100))
```

## Posição vertical em função do tempo

```
def calcPosiVelFunTemp(Szero, Vzero, grav, temp):
  vFinal1 = Szero + (Vzero * temp) + ((grav/2) * (temp ** 2))
  vFinal2 = Szero + (Vzero * temp) - ((grav/2) * (temp ** 2))
  return vFinal1, vFinal2
```

teste de serviço: posição vertical em função do tempo

```
print('Posição vertical em função do tempo', calcPosiVelFunTemp(50, 6, 13, 100))
```
